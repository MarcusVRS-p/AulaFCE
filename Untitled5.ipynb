```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
from scipy.sparse import diags
from scipy.sparse.linalg import spsolve
from scipy.sparse.linalg import bicgstab
from scipy.linalg import solve

def linalg_solver(temp_jm1, Ap, Ae, Aw, An, As, Afonte, TL1, TL2, max_iter=100, tol=1e-4):
    Nx, Ny = temp_jm1.shape
    T = temp_jm1.copy()

    # Create the coefficient matrix A and the constant vector b
    A = np.zeros((Nx * Ny, Nx * Ny))
    b = np.zeros(Nx * Ny)

    for i in range(Nx):
        for k in range(Ny):
            index = i * Ny + k
            ap = Ap[i, k]
            ae = Ae[i, k]
            aw = Aw[i, k]
            an = An[i, k] if k < Ny - 1 else 0
            ass = As[i, k] if k > 0 else 0

            A[index, index] = ap

            if i < Nx - 1:
                A[index, index + Ny] = -ae
            if i > 0:
                A[index, index - Ny] = -aw
            if k < Ny - 1:
                A[index, index + 1] = -an
            if k > 0:
                A[index, index - 1] = -ass

            b[index] = Afonte[i, k]

    # Solve the linear system using scipy.linalg.solve
    x = solve(A,b)

    # Reshape the solution vector back into the original matrix form
    T = x.reshape(Nx,Ny)

    return T

#entrada
alfa2=9.32e-6
alfa1=1.474e-5
k2=21.9
k1=54
Lx=0.1
Ly=0.04
T0=25
TL1=25
TL2=25
h1=10
hc2=0
hc1=1000

#malha
Nx=30
Ny=30
M=51
x=np.linspace(0,Lx,Nx)
dx=x[1]-x[0]
y=np.linspace(0,Ly,Ny)
dy=y[1]-y[0]
dt=30
q0=2e4

#------------alocacao de vetores e matrizes
temp=np.empty((M,Nx,Ny))
Ap=np.empty((Nx,Ny))
Ae=np.empty((Nx,Ny))
Aw=np.empty((Nx,Ny))
An=np.empty((Nx,Ny))
As=np.empty((Nx,Ny))
Afonte=np.empty((Nx,Ny))
temp[0,:,:]=T0*np.ones((Nx,Ny))
tempo=np.empty(M)
tempo[0]=0
#-------------------------------------- Constantes
A1=alfa1*dt/dx
A2=alfa2*dt/dx
B1=alfa1*dt/dy
B2=alfa2*dt/dy
C=1/dx
D=1/dy
E1=h1/k1
E2=h1/k2
F1=q0/k1
F2=q0/k2
O1=hc1/k1
O12=hc1/k2
O22=hc2/k2
##---------------------------------- Código
for j in range(1,M): # loop do tempo

    tempo[j]=dt*j ## t=k*dt

    for i in range(0,Nx):  # primeira linha
      Ap[i,0]=1+2*A1*C+B1*D+B1*E1 if 0<i<Nx-1 else 1+A1*C+B1*D+B1*E1
      Ae[i,0]=A1*C    if i<Nx-1 else 0
      Aw[i,0]=A1*C    if i>0 else 0
      An[i,0]=B1*D
      As[i,0]=0
      Afonte[i,0]=B1*E1*TL1 + temp[j-1,i,0]

#----------------------------------------------------------------#dominio
    for i in range(0,Nx):

        if tempo[j] >= int((M-1)*dt / 2) and i >= Nx / 2:
          hc_val = hc2
        else:
          hc_val = hc1

        for k in range(1,Ny-1):
          if k<int(Ny/2-1):
            Ap[i,k] = 1+2*A1*C+2*B1*D if 0<i<Nx-1 else 1+A1*C+2*B1*D
            Ae[i,k] = A1*C if i<Nx-1 else 0
            Aw[i,k] = A1*C if i>0 else 0
            An[i,k] = B1*D
            As[i,k] = B1*D
            Afonte[i,k] = temp[j-1,i,k]
          elif k==int(Ny/2-1):
            Ap[i,k] = k1/dy
            Ae[i,k] = 0
            Aw[i,k] = 0
            An[i,k] = hc_val
            As[i,k] = k1/dy-hc_val
            Afonte[i,k] = 0
          elif k==Ny/2:
            Ap[i,k] = k2/dy
            Ae[i,k] = 0
            Aw[i,k] = 0
            An[i,k] = k2/dy-hc_val
            As[i,k] = hc_val
            Afonte[i,k] = 0
          else:
            Ap[i,k] = 1+2*A2*C+2*B2*D if 0<i<Nx-1 else 1+A2*C+2*B2*D
            Ae[i,k] = A2*C if i<Nx-1 else 0
            Aw[i,k] = A2*C if i>0 else 0
            An[i,k] = B2*D
            As[i,k] = B2*D
            Afonte[i,k] = temp[j-1,i,k]

#----------------------------------------------------------------ultima linha

    for i in range(0,Nx):
        Ap[i,Ny-1]=1+2*A2*C+B2*D+B2*E2 if 0<i<Nx-1 else 1+A2*C+B2*D+B2*E2
        Ae[i,Ny-1]=A2*C if i<Nx-1 else 0
        Aw[i,Ny-1]=A2*C if i>0 else 0
        An[i,Ny-1]=0
        As[i,Ny-1]=B2*D
        Afonte[i,Ny-1]=F2*B2 + temp[j-1,i,Ny-1] + B2*E2*TL2


    temp[j,:,:] = linalg_solver(temp[j-1,:,:], Ap, Ae, Aw, An, As, Afonte, TL1, TL2)

###------------------------------- Valores comparação
y_value = [0.0000e+00, 4.3636e-03, 8.8081e-03, 1.3253e-02, 1.7697e-02,
     2.2141e-02, 2.6586e-02, 3.1030e-02, 3.5475e-02, 3.9919e-02]

T_value = [207.38, 207.38, 207.38, 207.87, 207.87,
     310.66, 311.15, 312.62, 315.57, 318.52]

# Debugging: Verificar os dados

print("Temperatura ao longo do tempo (última linha):", temp[:, 0, -1])
print("Temp final (última linha):", temp[-1, -1, :])

# Gráfico 1: Temperatura vs Posição (x)
plt.figure(1)
plt.plot(y, temp[-1, -1, :], label=f"t = {tempo[-1]} s")
plt.scatter(y_value, T_value, marker='o', color='r', label="Solução analítica")
plt.title("Temperatura vs Posição (y)")
plt.xlabel("Posição Y (m)")
plt.ylabel("Temperatura (°C)")
plt.grid(True)
plt.legend()
plt.show()

# Gráfico 2: Temperatura vs Tempo em uma posição específica
temperatura_tempo = temp[0, :, 0]

plt.figure(2)
plt.plot(x, temperatura_tempo)
plt.title("Temperatura vs Tempo")
plt.xlabel("Tempo (s)")
plt.ylabel("Temperatura (°C)")
plt.grid(True)
plt.legend()
plt.show()


from sklearn.metrics import mean_squared_error

# Interpolate the calculated temperature values to match the y_value points
from scipy.interpolate import interp1d

# Assuming the last time step contains the steady-state solution
calculated_temp_at_last_timestep = temp[-1, -1, :]
f = interp1d(y, calculated_temp_at_last_timestep, kind='linear')
interpolated_calculated_temps = f(y_value)

# Calculate the Mean Squared Error
mse = mean_squared_error(T_value, interpolated_calculated_temps)

print(f"Erro Quadrático Médio (MSE) entre a solução analítica e a calculada: {mse}")
```
